The iris.csv dataset, created by the statistician and biologist Ronald Fisher in 1936, contains measurements of three iris flower varieties: setosa, versicolor, and virginica. Each variety comprises 50 measurements of sepal length and width, as well as petal length and width. This dataset is renowned for its correlations among these measurements, facilitating the identification of iris varieties based solely on their relative proportions. In this exploration, we will delve into these correlations to gain insights into the relationships within the dataset.


1. Exploring the data

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# 1.Read the dataset from a CSV file into a pandas DataFrame
data = pd.read_csv('https://tinyurl.com/ChrisCoDV/iris.csv')

print(data.head())
print(data.tail())
print(data.describe())
print(data.sum())
```

2. Create a single visualization showing scatter plots for all pairs of measurements.

```python
selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

counter = 1
fig = plt.figure(figsize=(8, 8))
fig.suptitle('Measurements Correlations', fontsize=14, position=(0.5, 1.0))
for i, name_i in enumerate(selected):
  for j in range(i + 1, len(selected)):
    name_j = selected[j]
    sub = fig.add_subplot(3, 3, counter)
    sub.set_title(name_i + ' vs ' + name_j,fontsize=10)
    sub.scatter(data[name_i], data[name_j], s=0.5)
    counter += 1
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.tight_layout()
plt.show()
```

**Strongest Positive Correlations: sepal.length vs petal.width and petal.length vs petal.width**

3. Creating a heatmap illustrating the correlations among all the measurements.

```python
selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

plt.figure(figsize=(8, 8))
corr = data[selected].corr()
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(220, 20, n=200), square=True, annot=True, annot_kws={"size": 8})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

**Pairs of measurements that are most strongly correlated: petal.length vs sepal.length, petal.width vs sepal.length, petal.width vs petal.length.**

4. Next, let's explore correlations within the different iris varieties.

```python
data_setosa = data[data['variety'] == 'Setosa']
selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

counter = 1
fig = plt.figure(figsize=(8, 8))
fig.suptitle('Measurements Correlations', fontsize=14, position=(0.5, 1.0))
for i, name_i in enumerate(selected):
  for j in range(i + 1, len(selected)):
    name_j = selected[j]
    sub = fig.add_subplot(3, 3, counter)
    sub.set_title(name_i + ' vs ' + name_j,fontsize=10)
    sub.scatter(data_setosa[name_i], data_setosa[name_j], s=0.5)
    counter += 1
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.tight_layout()
plt.show()
```

**The strongest correlation for Setosa variety is sepal_length vs sepal_width.**

5. A heatmap provides a more precise measure than scatter subplots, so let's generate a heatmap illustrating the correlations between all pairs of measurements specifically for Setosa irises.

```python
data_setosa = data[data['variety'] == 'Setosa']
selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

plt.figure(figsize=(8, 8))
corr = data_setosa[selected].corr()
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(220, 20, n=200), square=True, annot=True, annot_kws={"size": 8})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

**The strongest positive correlation is sepal_length vs sepal_width. No evidence of inverse correlations**

7. Finally, let's summarize this into a single visualization using scatter plots, with each variety represented by a different color.

```python
selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

data_setosa = data[data['variety'] == 'Setosa']
data_versicolor = data[data['variety'] == 'Versicolor']
data_virginica = data[data['variety'] == 'Virginica']

counter = 1
fig = plt.figure(figsize=(8, 8))
fig.suptitle('Measurements Correlations', fontsize=14, position=(0.5, 1.0))
for i, name_i in enumerate(selected):
  for j in range(i + 1, len(selected)):
    name_j = selected[j]
    sub = fig.add_subplot(3, 3, counter)
    sub.set_title(name_i + ' vs ' + name_j,fontsize=10)
    sub.scatter(data_setosa[name_i], data_setosa[name_j], s=4)
    sub.scatter(data_versicolor[name_i], data_versicolor[name_j], s=4)
    sub.scatter(data_virginica[name_i], data_virginica[name_j], s=4)
    counter += 1
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.tight_layout()
plt.show()
```
