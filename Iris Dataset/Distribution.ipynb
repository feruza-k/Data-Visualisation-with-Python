
This week we will explore the distributions of the Iris dataset.

1. Exploring the data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1.Read the dataset from a CSV file into a pandas DataFrame
data = pd.read_csv('https://tinyurl.com/ChrisCoDV/iris.csv')

print(data.head())
print(data.tail())
print(data.describe())
print(data.sum())
```

2. Generate three separate box plot visualizations, one for each iris variety, illustrating the distribution of the four measurement types. Each visualization contains four box plots corresponding to sepal length, sepal width, petal length, and petal width.

```python
selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

data_setosa = data[data['variety'] == 'Setosa']
data_versicolor = data[data['variety'] == 'Versicolor']
data_virginica = data[data['variety'] == 'Virginica']

iris = [data_setosa, data_versicolor, data_virginica]

for i, name in enumerate(iris):  # Use enumerate to iterate over both index and value
    plt.figure(figsize=(8, 8))
    plt.boxplot(name[selected], labels=selected)
    plt.xlabel('Measurements (cm)', fontsize=18)
    plt.ylabel('Frequency', fontsize=18)
    plt.title(f'{iris[i].iloc[0]["variety"]} distributions', fontsize=20)
    plt.show()
```

**Versicolor and Virginica  varieties are most similar in terms of their distributions because of the spread of each measurement.**

3. Finally, let's generate three separate histogram visualizations, one for each iris variety, showcasing the distribution of the four measurement types, all superimposed on the same plot. This will result in three visualizations, each containing four histograms corresponding to sepal length, sepal width, petal length, and petal width.

```python
selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

data_setosa = data[data['variety'] == 'Setosa']
data_versicolor = data[data['variety'] == 'Versicolor']
data_virginica = data[data['variety'] == 'Virginica']

iris = [data_setosa, data_versicolor, data_virginica]

x_min = 0
x_max = 8
bin_width = 0.1
# the bin calculations
n_bins = int((bin_width + x_max - x_min) / bin_width)
print(str(n_bins) + ' bins')
bins = [(x_min + x * (bin_width + x_max - x_min) /(n_bins + 1)) for x in range(int(n_bins + 1))]


for i, name in enumerate(iris):
    for label in selected:
        plt.hist(name[label], bins, alpha=0.5, label=label)
    plt.title('Iris ' + name.iloc[0]['variety'] + ' Measurement Distributions', fontsize=20)
    plt.xlabel('Measurements (cm)', fontsize=18)
    plt.ylabel('Frequency', fontsize=18)
    plt.legend()
    plt.show()
```
